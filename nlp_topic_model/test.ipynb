```python
# Here are the sample documents combining together to form a corpus.

doc1 = "Sugar is bad to consume. My sister likes to have sugar, but not my father. bad"

doc2 = "My father spends a lot of time driving my sister around to dance practice. sister driving driving"

doc3 = "Doctors suggest that driving may cause increased stress and blood pressure. Doctors stress"

doc4 = "Sometimes I feel pressure to perform well at school, but my father never seems to drive my sister to do better. pressure"

doc5 = "Health experts say that Sugar is not good for your lifestyle. Health lifestyle"
```

```python
# compile documents

doc_list = [doc1, doc2, doc3, doc4, doc5]
doc_list
```

```python
type(doc_list)
```

```python
# List to Dataframe
import pandas as pd
df = pd.DataFrame(doc_list, columns =['text']) 
df
```

```python
# Import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')

# Create Document Term Matrix
dtm_csr = cv.fit_transform(df['text'])
```

```python
dtm_df = pd.DataFrame(dtm_csr.toarray(), columns=cv.get_feature_names())
dtm_df.index = df.index
dtm_df.head()
```
